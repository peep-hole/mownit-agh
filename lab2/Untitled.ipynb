```python
import numpy as np
```

```python
def gauss_jordan(Y,X):
    n = Y.shape[0]
    for i in range(n):
        pivot_index = i
        for j in range(i+1,n):
            if abs(Y[j][i]) > abs(Y[pivot_index][i]):
                pivot_index = j
        Y[i], Y[pivot_index] = Y[pivot_index],Y[i]
        X[i], X[pivot_index] = X[pivot_index], X[i]
        Y[i, :] /= Y[i][i]
        X[i] /= Y[i][i]
        for j in range(n):
            if j != i:
                Y[j] = Y[j] - Y[i,:] * Y[j][i]
                X[j] = X[j] - X[i] * Y[j][i]       
    return X



A = np.array([[1.0, 2.0, 3.0], [1.0, 7.0, 2.0], [1.0, 5.0, 2.0]])
B = np.array([[1.0], [0.0], [5.0]])

res = gauss_jordan(A, B)
print("Gauss: \n",res)
print('\n', "linalg: \n", np.linalg.solve(A,B))
                
            
```

```python
def LU(A):
    n = A.shape[0]
    U = A.copy()
    L=np.eye(n)
    for i in range(n):
        f = U[i+1:, i] / U[i, i]
        L[i+1:, i] = f
        U[i+1:] -= f[:, np.newaxis] * U[i]
    return L, U    
    
C = np.array([[1.0, 2.0, 3.0], [1.0, 7.0, 2.0], [1.0, 5.0, 2.0]])   

lu_c = LU(C)
print(lu_c[0])
print(lu_c[1])
```
